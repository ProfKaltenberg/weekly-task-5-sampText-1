```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
path='C:/Users/kanik/OneDrive/Documents/ECO 590/weekly-task-5-sampText-1/weekly task 5/'
city=pd.read_csv(path+'us-cities-top-1k-multi-year.csv')
city
```

```python
pop = city[['year','State','Population']].drop_duplicates().groupby(['State','year'])['Population'].sum().reset_index()
# include ['Population'] on the outside of the groupby so all the population values are added up with .sum()
question=pop[(pop['State']=='New York')|(pop['State']=='New Jersey')|(pop['State']=='Connecticut')|(pop['State']=='Pennsylvania')]
question
```

```python
state_unique=question['State'].unique()
# makes a list of the four states we want
colors=['red','blue','green','purple']
# list of colors to be used
c=0

for s in state_unique:
    state=question[(question['State']==str(s))]
    # checks the dataframe for matches to the each state
    plt.plot(state['year'],state['Population'],label=s,color=colors[c])
    # plots the values
    c+=1

plt.legend(loc='upper left',bbox_to_anchor=(1.05, 1))
plt.xticks([i for i in range(2014,2019,1)])

plt.xlabel('Year', fontweight='bold')
plt.ylabel('Population in 10s of Millions',fontweight='bold')
plt.title("State Population Over Time",fontweight='bold')
```

# Question 2

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a Bar Graph of the total population for 5 cities in the state of New York comparing 2014 years and 2018 years (15 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Combine the graph in question 1 and question 2 in one figure.

```python
path='C:/Users/kanik/OneDrive/Documents/ECO 590/weekly-task-5-sampText-1/weekly task 5/'
city=pd.read_csv(path+'us-cities-top-1k-multi-year.csv')
city
```

```python
state = city[['year','State','City','Population']].drop_duplicates().groupby(['State','City','year','Population']).count().reset_index()

ny_cities=['Albany','Buffalo','Rochester','Syracuse','Utica']

ny=state[(state['State']=='New York')&(state['City'].isin(ny_cities))&((state['year']==2014)|(state['year']==2018))]
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html
ny=ny.reset_index().drop('index',axis=1)
ny
```

```python
state_pop=ny[['year','Population']].groupby(['year'])['Population'].sum().reset_index()
state_pop['year']=state_pop['year'].astype(str)
plt.bar(state_pop['year'],state_pop['Population'])

plt.xlabel("Year",fontweight='bold')
plt.ylabel("Population in Millions",fontweight='bold')
plt.title("NY Population",fontweight='bold')
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax1 = plt.subplot(2,1,1)
ax2 = plt.subplot(2,1,2)


ax1.hist(job['prob'],bins=40)
ax1.set_xlabel('Probability of Automation',fontweight='bold')
ax1.set_title('Histogram of Probability',fontweight='bold')

ax2.plot('Average annual wage','numbEmployed',data=job,linestyle='none',marker='.')
ax2.set_xlabel('Average Annual Wage in Millions of $',fontweight='bold')
ax2.set_ylabel('Number of Employed',fontweight='bold')
ax2.set_yscale('log')
# https://matplotlib.org/stable/users/explain/axes/axes_scales.html
ax2.set_title('# of Employed vs Average Annual Wage',fontweight='bold')

fig.tight_layout()
fig.savefig('Weekly_Task_5.png', dpi = 300)
```
